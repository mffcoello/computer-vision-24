```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
```

```python
def scrapy(query):
    url = f"https://www.ebay.com/sch/i.html?_nkw={urllib.parse.quote_plus(query)}"

    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        items = soup.select('.s-item')

        for index, item in enumerate(items):
            title = item.select_one('.s-item__title').text.strip()
            price = item.select_one('.s-item__price').text.strip()

            try:
                rating = item.select_one('.s-item__reviews').text.strip()
            except:
                rating = "Rating not available"

            item_url = item.select_one('.s-item__link')['href']

            print(f"\nItem {index + 1}:")
            print(f"Title: {title}")
            print(f"Rating: {rating}")
            print(f"Price: {price}")
            print(f"Item URL: {item_url}")

    except Exception as e:
        print(f"Error extracting product details: {e}")

def main():
    query = input("Enter the product name to search for: ")
    scrapy(query)

    input("\nPress Enter to exit the program")
    print()

if __name__ == "__main__":
    main()
```
